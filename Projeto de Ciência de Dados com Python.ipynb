### Exercício 4

Buscando os IDs duplicados no DataFrame original. Para isso, criamos uma máscara booleana `dupe_mask` que receberá todos os IDs do DataFrame original e marcará como `True` aqueles que possuem 2 entradas no DataFrame original e como `False` todas as outras. Com essa máscara booleana iremos criar uma lista `dupe_ids` que contém todos os IDs duplicados.

```python
dupe_mask = (id_counts == 2)
dupe_mask
```

```python
dupe_ids = list(id_counts.index[dupe_mask])
print(dupe_ids[:5])
print(len(dupe_ids))
```

Este comando é um pouco mais complexo. O objetivo dele é usar a lista dupe_ids para fazer uma busca no DataFrame original
para trazer as linhas cujo ID está duplicado. O comando df.loc realiza a busca baseada nos parâmetros passados.
Esses parâmetros, por sua vez, são `df['ID']`, uma vez que estamos usando o ID como ponto de filtragem, `.isin(duple_ids)` para buscar somente os elementos que fazem parte da lista `duple_ids`, e o `sort_values('ID')` pois queremos ordenar a nova tabela com base nos IDs para que IDs iguais fiquem juntos, facilitando assim a leitura.

Podemos ver que a tabela final possui 626 elementos. Esse número era esperado pois, como vimos acima, o DataFrame original possui 313 IDs duplicados.

```python
df.loc[df['ID'].isin(dupe_ids)].sort_values('ID')
```

Vemos então que todas as contas duplicadas possuem uma linha com valores reais e uma com valores igual a 0 para todas as features. Essas segundas linhas podem (e devem) ser excluídas, uma vez que elas não possuem qualquer significado real.

```python
df_zero_mask = (df == 0)
df_zero_mask
```

O DataFrame acima é uma máscara booleana onde todos os valores igual a 0 serão marcados como True. Isso não é bastante, uma vez que alguns valores realmente podem ser iguais a 0 mesmo para linhas válidas. Importante notar também que não haverá nenhuma linha na coluna ID marcada como True, uma vez que não é possível existis ID 0.

O próximo passo é filtrar, a partir dessa tabela, as linhas que possuem 0 para todas as colunas (exceto ID) e removê-las. Uma opção é criar a `feature_zero_mask`. O que acontece nessa feature é uma seleção `.iloc` do DataFrame `df_zero_mask`. Para o `.iloc` é necessário passar os parâmetros de seleção. Nesse caso temos `[:, 1:]`. Aqui estamos fazendo duas filtragrens. A primeira, marcada somente pelo `:`, delimita que todas as linhas serão analisadas. A segunda, `1:`, delimita que só serão analisadas as colunas a partir da indexada pelo índice 1 (ou seja, a segunda, `LIMIT_BAL`) até o final. Por fim, o método `.all(axis = 1)` nos mostra uma lista simples caso todas as colunas (`axis = 1`) sejam True, exceto a primeira, como já foi especificado anteriormente. Dessa forma, todas as linhas marcadas como True nesse último processo podem ser consideradas linhas duplicadas com valores igual a 0 para todas as features, exceto ID

```python
feature_zero_mask = df_zero_mask.iloc[:, 1:].all(axis = 1)
feature_zero_mask
```

```python
sum(feature_zero_mask)
```

```python
df_clean = df.loc[~ feature_zero_mask, :].copy()
```

```python
df_clean.shape[0] == df_clean['ID'].nunique()
```

Agora podemos ver que o tamanho do DataFrame `df_clean` é igual o número de IDs únicos. Ou seja, nosso DataFrame está limpo e organizado.
